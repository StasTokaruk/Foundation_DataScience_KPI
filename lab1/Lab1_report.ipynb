```python
import pandas as pd
```

```python
# Завантаження Dataset з платформи kaggle
import kagglehub

# Download latest version
path = kagglehub.dataset_download("shivamb/amazon-prime-movies-and-tv-shows")

print("Path to dataset files:", path)
```

```python
# 
df = pd.read_csv('amazon_prime_titles.csv')
df
```

```python
# Вивід загального опису параметрів стовпця "country"
df.country.describe()
```

```python
# Вивід унікальних країн та їх комбінацій
df.country.unique()
```

```python
# Вивід середнього року випуску фільмів
df.release_year.mean()
```

```python
# Вивід кількості випущених фільмів та серіалів країною
df.country.value_counts()
```

```python
# Вивід відфільтрованих за типом продакшинів
movies = df[df['type'] == 'Movie'].copy()
movies
```

```python
# Вивід середньго значення тривалості фільму
movies_duration_int = movies['duration'].str.replace(" min", "").astype(int)
duration_movies_mean = movies_duration_int.mean()
duration_movies_mean
```

```python
# Вивід відхилення від сердньої тривалості фільму
movies['difference_mean'] = movies_duration_int - duration_movies_mean
movies[['title', 'difference_mean']]
```

```python
# Вивід назви фільму з його триваістю
df[['title', 'duration']]
```

```python
# Вивід відфільтрованих за жанром фільмів
df[df['listed_in'].str.contains(r"\bcomedy\b", case=False)]
```

```python
# Кількість років, що пройшли від релізу фільму
df['passed_years'] = df.release_year.map(lambda p : 2025 - p)
df[['title', 'passed_years']]
```
