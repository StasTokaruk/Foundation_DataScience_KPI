 ## Четвертий етап аналізу даних
### Обробка пропущених даних

#### Приклади обробки пропущених даних:
#### 1) Видалення рядків або стовпців з пропусками dropna()
##### Цей підхід не рекомендується, оскільки може призвести до втрати важливої інформації, тому мені він не підходить.
#### a) Видалення рядків

```python
# Видаляємо всі рядки з NaN
drop_rows = movie_dat.dropna()
drop_rows
```

#### b) Видалення стовпців

```python
# Видаляємо всі стовпці з NaN
drop_cols = movie_dat.dropna(axis=1)
drop_cols
```

#### 2) Заповнення набору даних певним значенням fillna(value)
##### Цей підхід вже кращий, але в моєму DataFrame немає закономірності для заповнення, тому дані потрібно витягувати з інших джерел.


```python
# Приклад заповнення нулями.
df_filled = movie_dat.fillna(0)
df_filled
```

```python
#Можна заповнювати тільки певну частину DataFrame (наприклад, один стовпець):
df_filled_director = movie_dat['director'].fillna("Unknown")
df_filled_director
```

#### 3) Заповнення наступним або попереднім значенням у стовпці ffill / bfill.
##### Цей метод корисний для послідовних спостережень (наприклад, часові ряди), коли пропуск логічно замінювати сусіднім значенням. 

```python
# заповнює пропуск значенням з попереднього рядка.
df_ffill = movie_dat.ffill()
df_ffill
```

```python
# заповнює пропуск значенням з наступного рядка.
df_bfill = movie_dat.bfill()
df_bfill
```

### Заповнення даних про режисерів за допомогою OMDb API.

```python
API_KEY = "177265b7"  

for idr, row in movie_dat[movie_dat['director'].isnull()].iterrows():
     title = row['title']
     url = f"http://www.omdbapi.com/?t={title}&apikey={API_KEY}"
     
     response = requests.get(url).json()
     
     if response.get('Response') == 'True':
         if pd.isnull(movie_dat.at[idr, 'director']):
            movie_dat.at[idr, 'director'] = response.get('Director', movie_dat.at[idr, 'director'])
         
movie_dat
```
