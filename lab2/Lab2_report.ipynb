### Індексно-орієнтований вибір

```python
# Вибрали всі поля першого рядка
movie_dat.iloc[0]
```

```python
# Вибір стовпця на основі індексу
movie_dat.iloc[:, 2]
```

```python
#Вибір стовпця на основі індексу з одночасною операцією зрізу з елементів стовпця в Series.
movie_dat.iloc[1:4, 2]
```

### 3. Мітко-орієнтований вибір в бібліотеці Pandas

```python
# Вибір по індексу елементу стовпця заданого по імені.
movie_dat.loc[0, 'title']
```

```python
# Вибір підмножини стовпців, імена яких задані списком
movie_dat.loc[:, ['type', 'title', 'release_year']]
```

```python
#Вибір перших 10 рядків з підмножини стовпців, імена яких задані списком
movie_dat.loc[:11, ['type', 'title', 'release_year']]
```

```python
#Вибір рядків з кроком у заданому діапазоні з підмножини стовпців, імена яких задані списком
movie_dat.loc[:50:10, ['type', 'title', 'release_year']]
```

### 4. Маніпулювання індексом

```python
movie_dat.set_index('title')
```

### 4. Умовний вибір

```python
# Проста умова
# Наприклад, ми хочемо сформувати піднабір даних, який
# складається з фільмів, дата релізу яких є більшою за середню.
ry_mean = movie_dat.release_year.mean()
print(f"Середній рік релізу фільмів {round(ry_mean)}")
movie_dat.loc[movie_dat.release_year > ry_mean]
```

```python
# Складна умова
# Складна умова дозволяє виконувати логічні операції вибору з використанням логічних операторів «&» та «|».
movie_dat.loc[(movie_dat.release_year > ry_mean) & movie_dat.listed_in.str.contains('Comedy')]
```

```python
#Умовні селектори
movie_dat.loc[movie_dat.listed_in.isin(['Comedy'])]
```

```python
#Умовні селектори кілька елементів одночасно
movie_dat.loc[movie_dat.listed_in.isin(['Comedy', 'Drama'])]
```

```python
#Умовні селектори, перевірка порожніх (або не порожніх) значень.
movie_dat.loc[movie_dat.country.isnull()]
```

```python
movie_dat.loc[movie_dat.country.notnull()]
```
